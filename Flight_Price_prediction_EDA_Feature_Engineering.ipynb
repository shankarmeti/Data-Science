# Arrival_Time feature 

in this feature we need only 01:10 

22 mar not required 

so that by using split function we will split 01:22 and 22 mar 

```python
final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
#using Lambda function
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
# now we will split Arrival_Time into two features 1st Arrival_Hour and 2nd Arrival_Minutes

final_df['Arrival_Hour']=final_df['Arrival_Time'].apply(lambda x : x.split(':')[0])
final_df['Arrival_Minutes'] = final_df['Arrival_Time'].apply(lambda x:x.split(':')[1])
```

```python
final_df.head()
```

```python
# its still in object type now we need to convert into interger type

final_df[['Arrival_Hour','Arrival_Minutes']]=final_df[['Arrival_Hour','Arrival_Minutes']].astype(int)
```

```python
final_df.info()
```

```python
# now Arrival_Time  is not required, so that we need to drop that column

final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(1)
```

# Dep_Time feature

 Dep_Time : 22:20
 
 so we need to split the Dep_Time in two features 1st Dep_Hour, 2nd Dep_Minutes
 
 after creating two features we need to delete Dep_Time

```python
final_df['Dep_Hour']=final_df['Dep_Time'].apply(lambda x : x.split(':')[0])
final_df['Dep_Minutes'] = final_df['Dep_Time'].apply(lambda x : x.split(':')[1])
```

```python
final_df.head(1)
```

```python
#now we need to convert dtype to int
final_df[['Dep_Hour','Dep_Minutes']]=final_df[['Dep_Hour','Dep_Minutes']].astype(int)
```

```python
final_df.info()
```

```python
#drop Dep_Time from final_df

final_df.drop('Dep_Time',axis=1,inplace=True)
```

# Total_Stops feature

Total_Stops has values like non-stop , 1 stop , 2 stop , 3 stop and nan, 

so that we can easily replace those values with dummies 

ex: non-stop = 0

1 stop = 1

2 stop = 2

3 stop = 3

by using map function we can replace those values with numbers

```python
# 1st we need to check unique values to replace
final_df['Total_Stops'].unique()
```

```python
# replacement using map() function
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'NaN':1})
```

```python
final_df.head()
```

```python
# we have one missing value in Total_Stops
final_df[final_df['Total_Stops'].isnull()]
```

```python
#to fill the missing value we need to check mode, beacuse which value is reapating we can easily fill with that value in place of missing value
mode=final_df['Total_Stops'].mode()
```

# Airline Feature

```python
#now we need check unique values of Airline column

final_df['Airline'].unique()
```

```python
#now we will give labels to that values
#imoprt sklearn for label encoding

from sklearn.preprocessing import LabelEncoder
labelencode=LabelEncoder()
```

```python
#now we will give labels to remaining categorical features

#1)Airline
#2)Source
#3)Destination
#4)Additional_Info

final_df['Airline']=labelencode.fit_transform(final_df['Airline'])
final_df['Source']=labelencode.fit_transform(final_df['Airline'])
final_df['Destination']=labelencode.fit_transform(final_df['Airline'])
final_df['Additional_Info']=labelencode.fit_transform(final_df['Airline'])
```
