```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
# country code is column in df zomato dataset
df_country
```

# now we need to combine two dataframes (df & df_country)

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')

# on = mention comman column in both dataset 

# how = left join or right join 
```

```python
final_df.head(2)
```

```python
#  To Check data types
final_df.dtypes
```

```python
# im trying to find out how many country are there and records/country
final_df.Country.value_counts()

# from this zomato is mostly available in india
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
## pie chart - Top 3 coutry that uses zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

Observation :Zomato maximum records or transaction are from India

After that USA and Then UK

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
    
    # renaming the column name
    #reset.index() = it coverts the output in to dataframe
```

```python
ratings
```

# Observation
1. When Rating is between 4.5 to 4.9 -----> Excellent

2. When Rating is between 4.4 to 4.0 -----> Very Good

3.When Rating is between 3.9 to 3.5 -----> Good

4.When Rating is between 3.4 to 2.5 -----> Avarage
etc


```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
plt.rcParams['figure.figsize']=(15,6)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings)
#matching the colors with Rating color
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])

# here we can match exact colors
```

# Observation: 
    1. Not Rated count is very hogh
    2.max numbers of rationg betwwen 2.5 to 3.4

```python
## count plot : plot wrt categorical variables

sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```
